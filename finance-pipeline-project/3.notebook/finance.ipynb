```python
import pyodbc
import pandas as pd
from sqlalchemy import create_engine
```

```python
# 1. Leer el CSV
df = pd.read_csv(r'C:\Users\Ari\Desktop\DATA SET\finance-pipeline-project\1.data\financial_transactions.csv', encoding="utf-8")
```

```python
# 2. Conexión a SQL Server
engine = create_engine(
    "mssql+pyodbc://localhost/FinancialPipelineDB"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes" )
```

```python
# 3. Cargar a SQL
df.to_sql(
    name="transactions",
    con=engine,
    if_exists="append",
    index=False
)

print("Carga completada")
```
